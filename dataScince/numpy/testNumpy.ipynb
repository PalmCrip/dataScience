```python
#创建一个3*2的数组
import numpy as np
arr=np.array([[1,2],[3,4],[5,6]])
print("形状shape",arr.shape)
print("纬度ndim：",arr.ndim)
print("总元素",arr.size)
print("元素类型dtype:",arr.dtype)
print("转置T:",arr)
```

```python
#数组内存计算
arr = np.array([[0, 1], [2, 3]], dtype=np.int32)
print("内存",arr.itemsize)
print("内存",arr.nbytes)
```

```python
### np.lineSpace：等间隔数组
import numpy as np
### zeros_like(): 返回与给定数据具有相同形状和类型的0新书组
### ones_like()：返回与给定数组具有相同形状和类型的1新数组。
###  empty_like()：返回与给定数组具有相同形状和类型的未初始化的新数组。

arr2 = np.ones_like(arr, dtype=int)
arr4 = np.empty_like(arr2)
arr2 = np.full_like(arr2,5)
```
